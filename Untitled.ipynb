```python
def get_data_from_database(trans_id, sensor_data=True):
    '''
    该函数从数据库中查询发送节点id 对应的发送点坐标，以及若干个感知节点的接收信号。
    Parameters
        ----------
        trans_id: 发射节点的id
        sensor_data: 是否模拟感知的数据， Ture则返回100个感知节点的数据，False则返回所有数据500×500
    Returns
        ----------
        data_dict: 字典格式的数据，key如下
          trans_x: 发射节点x坐标，float
          trans_y: 发射节点y坐标，float
          recv_xs: 接收点的x坐标，NumpyArray
          recv_ys: 接收点的y坐标，NumpyArray
          tran_recv_distance: 发射点与每个接收点的距离，NumpyArray
          shadow_loss: 阴影衰落值，NumpyArray     
    '''
    data_dict = {}
    
    conn = psycopg2.connect(host="47.111.23.160", dbname="research", user="nicegeek", password="iloveting1314")
    cur = conn.cursor()
    
    # 查询发送节点的坐标
    cur.execute("""
    SELECT x_cord,y_cord FROM thesis1.transmission WHERE id=%(trans_id)s
    """, {'trans_id':trans_id})
    rows = cur.fetchone()
    data_dict['trans_x'], data_dict['trans_y'] = rows
    
    # 查询某一发送节点对应的 空间所有点的接收数据或者 感知数据
    cur.execute("""
           SELECT x_cord, y_cord, trans_id, tran_recv_distance, shadow_loss
        FROM thesis1.all_channel_lines_shadow
        INNER JOIN thesis1.all_points
        ON recv_id=thesis1.all_points.id
        WHERE trans_id=%(trans_id)s""", {'trans_id':trans_id})
    rows = cur.fetchall()
    data_ = pd.DataFrame(rows, columns=[x.name for x in cur.description])
    cur.close()
    conn.close()    
    data_=data_.sort_values(by=['y_cord','x_cord'], ascending=[False,True])
    if sensor_data is True:
        x = np.arange(24.5, 500, 50)
        y = np.arange(-474.5, 0, 50)
        data_ = data_[data_.x_cord.isin(x) & data_.y_cord.isin(y)]
    data_dict['recv_xs'] = data_['x_cord'].values
    data_dict['recv_ys'] = data_['y_cord'].values
    data_dict['tran_recv_distances'] = data_['tran_recv_distance'].values
    data_dict['shadow_loss'] = data_['shadow_loss'].values
    
    return data_dict

```

```python
def add_trans_model(data_dict, trans_power=50, η=3.5, m_error=False, error_theta=1):
    '''
    计算每个节点的接收功率值
    '''
    if m_error is True:
        error = np.sqrt(error_theta) * np.random.randn(len(data_dict['shadow_loss']))
        data_dict['recv_power']=trans_power - 10 * η * np.log10(data_dict['tran_recv_distances']) + data_dict['shadow_loss'] + error
    data_dict['recv_power']=trans_power - 10 * η * np.log10(data_dict['tran_recv_distances']) + data_dict['shadow_loss']
    data_dict['trans_power'] = trans_power
 #  trans_x: 发射节点x坐标，float
#          trans_y: 发射节点y坐标，float
#          recv_xs: 接收点的x坐标，NumpyArray
#          recv_ys: 接收点的y坐标，NumpyArray
#          tran_recv_distance: 发射点与每个接收点的距离，NumpyArray
#          shadow_loss: 阴影衰落值，NumpyArray
# trans_power:
#recv_power   
```

```python
def get_data_list(trans_id_list, trans_power=50, η=3.5, m_error=False, error_theta=1):
    data=[]
    for id in trans_id_list:
        each_group = get_data_from_database(id, sensor_data=True)
        add_trans_model(each_group, trans_power=trans_power, η=η, m_error=m_error, error_theta=error_theta)
        data.append(each_group)
    return data
```

```python
def stan_data_gen(data_list):
    stan_data = {}
    stan_data['D'] = len(data_list)
    stan_data['N'] = len(data_list[0]['recv_power'])
    stan_data['recv_x'] = list(data_list[0]['recv_xs'])
    stan_data['recv_y'] = list(data_list[0]['recv_ys'])
    
    stan_data['tran_power'] = []
    stan_data['tran_x'] = []
    stan_data['tran_y'] = []
    stan_data['tran_recv_distance'] = []
    stan_data['Y'] = []
    
    for each_group in data_list:
        stan_data['tran_power'].append(each_group['trans_power'])
        stan_data['tran_x'].append(each_group['trans_x'])
        stan_data['tran_y'].append(each_group['trans_y'])
        stan_data['tran_recv_distance'].append(list(each_group['tran_recv_distances']))
        stan_data['Y'].append(list(each_group['recv_power']))
        
    return stan_data
```

```python
%%stan exponential_cov
functions {
  matrix cov_exponential(matrix distance_between, real sigma, real fai) {
    int N=rows(distance_between);
    matrix[N, N] cov;
    for (i in 1:N){
      for (j in 1:N){
        if (i==j)
          cov[i, j] = 1;
        else
          cov[i,j] = square(sigma) * exp(-1*fai*distance_between[i,j]);
      }
    }
    
    return cov;
  }
}

data {
  int<lower=1> D; //数据组数
  int<lower=1> N; //感知节点个数
  vector[N] recv_x;// 每组数据有N个感知节点
  vector[N] recv_y;// 
  real tran_power[D]; //信号源发送功率
  real tran_x[D]; //信号源坐标
  real tran_y[D];
  vector[N] tran_recv_distance[D];
  vector[N] Y[D];
 
}

transformed data {
  matrix[N, N] distance_between_recv;
  for (i in 1:N){
    for (j in 1:N){
      distance_between_recv[i,j] = sqrt((recv_x[i]-recv_x[j])^2 + (recv_y[i]-recv_y[j])^2);
    }
  }
}

parameters {
  real gama; // 自由空间传播损耗因子
  real fai;  // matern 协方差函数的距离 参数
  real sigma; // 空间方差  
  real sigma_eta; // 测量误差
}
transformed parameters {

}

model {
  matrix[N,N] bigSigma;
  vector[N] i_vec = rep_vector(1.0, N);
  matrix[N, N] unit_matrix = diag_matrix(i_vec);
  
  /********************/
  /*    指定先验分布    */
  /********************/
  gama ~ uniform(1,10);
  fai ~ normal(20, 100);
  sigma ~ inv_gamma(2, 10);
  sigma_eta ~ inv_gamma(2, 10);
 
  
  
  bigSigma = cov_exponential(distance_between_recv, sigma, fai) + square(sigma_eta)*unit_matrix;
  for (i in 1:D){
    Y[i] ~ multi_normal(tran_power[i] - 10 * gama * log10(tran_recv_distance[i]), bigSigma); 
  }
   
}
```

```python
%%stan spectrum_nn
functions {
  matrix cov_matern(matrix x, real sigma, real fai, int upsilon) {
    int N=rows(x);
    matrix[N, N] cov;
    for (i in 1:N){
      for (j in 1:N){
        if (i==j)
          cov[i, j] = 1;
        else
          cov[i,j] = ((sigma ^ 2) / (2^(upsilon - 1) * tgamma(upsilon)))
            *((fai * x[i,j]) ^ upsilon) * modified_bessel_second_kind(upsilon, fai * x[i,j]);
      }
    }
    
    return cov;
  }
}

data {
  int<lower=1> D; //数据组数
  int<lower=1> N; //感知节点个数
  vector[N] recv_x;// 每组数据有N个感知节点
  vector[N] recv_y;// 
  real tran_power[D]; //信号源发送功率
  real tran_x[D]; //信号源坐标
  real tran_y[D];
  vector[N] tran_recv_distance[D];
  vector[N] Y[D];
 
}

transformed data {
  matrix[N, N] distance_between_recv;
  for (i in 1:N){
    for (j in 1:N){
      distance_between_recv[i,j] = sqrt((recv_x[i]-recv_x[j])^2 + (recv_y[i]-recv_y[j])^2);
    }
  }
}

parameters {
  real gama; // 自由空间传播损耗因子
  real fai;  // matern 协方差函数的距离 参数
  real sigma; // 空间方差  
  real sigma_eta; // 测量误差
}
transformed parameters {

}

model {
  matrix[N,N] bigSigma;
  vector[N] i_vec = rep_vector(1.0, N);
  matrix[N, N] unit_matrix = diag_matrix(i_vec);
  int upsilon = 2; 
  
  /********************/
  /*    指定先验分布    */
  /********************/
  gama ~ uniform(1,10);
  fai ~ uniform(1, 100);
  sigma ~ inv_gamma(2, 10);
  sigma_eta ~ inv_gamma(2, 10);
 
  
  
  bigSigma = cov_matern(distance_between_recv, sigma, fai, upsilon) + sigma_eta^2*unit_matrix;
  for (i in 1:D){
    Y[i] ~ multi_normal(tran_power[i] - 10 * gama * log10(tran_recv_distance[i]), bigSigma); 
  }
   
}
```
